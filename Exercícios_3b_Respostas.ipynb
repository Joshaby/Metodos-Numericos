```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
def arctan(x, n) :
    return ((-1) ** n * x ** (2 * n + 1)) / (2 * n + 1)

x = np.linspace(1, 19, 19)
fM = []
fH = []
fC = []
fD = []
auxM = 0
auxH = 0
auxC = 0
auxD = 0
for i in range(20) :
    auxM += ((4 * arctan(1 / 5, i) - arctan(1 / 239, i)) * 4)
    auxH += ((arctan(1 / 2, i) + arctan(1 / 3, i)) * 4)
    auxC += ((2 * arctan(1 / 3, i) + arctan(1 / 7, i)) * 4)
    auxD += ((arctan(1 / 2, i) + arctan(1 / 5, i) + arctan(1 / 8, i)) * 4)
    fM.append(auxM)
    fH.append(auxH)
    fC.append(auxC)
    fD.append(auxD)
plt.figure(figsize = (25, 10))
plt.plot(x, np.abs(np.diff(fM) / fM[0:19]), 'k', label = "F√≥rmula de Machin", c = "r")
plt.plot(x, np.abs(np.diff(fH) / fH[0:19]), 'k', label = "F√≥rmula de Hutton", c = "g")
plt.plot(x, np.abs(np.diff(fC) / fC[0:19]), 'k', label = "F√≥rmula de Clausen", c = "b")
plt.plot(x, np.abs(np.diff(fD) / fD[0:19]), 'k', label = "F√≥rmula de Dase", c = "y")
plt.grid()
plt.xticks(x, x)
plt.ylabel("Erro")
plt.xlabel("Termos")
plt.legend()
```

```python
eM = np.abs(np.diff(fM) / fM[:-1])
eH = np.abs(np.diff(fH) / fH[:-1])
eC = np.abs(np.diff(fC) / fC[:-1])
eD = np.abs(np.diff(fD) / fD[:-1])
print("F√≥rmula de Machin", fM[-1])
print("F√≥rmula de Hutton", fH[-1])
print("F√≥rmula de Clausen", fC[-1])
print("F√≥rmula de Dase", fD[-1])
print("PI do NumPy", np.pi)
print(eM)
print(eH)
print(eC)
print(eD)
```
