Exercício 1.1

```python
def divisaoEMedia(num) :
    l = list()
    for i in range(10) :
        if (i == 0) :
            x = 1
        else :
            x = l[i - 1]
        l.append((x + num / x) / 2)
    return l

def ErroRelativo(num) :
    l = divisaoEMedia(num)
    for i in l :
        er = np.abs(1 - i / num ** (1 / 2))
        print("Erro relativo x = %5.5f e ~x = %.5f = %.5f" %(num ** (1 / 2), i, er))
```

```python
ErroRelativo(36)
```

Exercício 1.2

```python
def eps() :
    e = 1
    while (not(1 + e <= 1)) :
        e /= 2
    e *= 2
    return e
```

```python
print(eps())
print(np.finfo(dtype=np.float64).eps)
```

```python
eps() == np.finfo(dtype=np.float64).eps
```

Exercício 1.4

```python
def exp1(mi) :
    return np.exp(1 / mi) / (1 + np.exp(1 / mi))
def exp2(mi) :
    return 1 / (np.exp(-1 / mi) + 1)
```

```python
a = 0.0001
print(exp1(a))
print(exp2(a))
```

Expressão 2 é mais correta! Porque na primeira expressão ocorre um overflow nos np.exp(1 / mi) depois de um determinado número muito pequeno! Por exemplo, np.exp(1 / 0.001) == 1.970071114×10$^{434}$ que é maior do que 1.7976931348623157x10$^{308}$ que é o limite de um float64 usado pelo python! 

Exercício 1.5

```python
def identidade(x) :
    return ((1 + x) - 1) / x
```

```python
print(identidade(1e-13))
print(identidade(1e-16))
print(identidade(1e-18))
```

Acabou ocorrendo um underflow, pois o número usado no cálculo fica cada vez menor, menor do que o menor valor que a máquina consegue operar, e logo, é retornado um 0! Ao longo das operações, os resultados ficam próximos do valor esperado que é 1!
