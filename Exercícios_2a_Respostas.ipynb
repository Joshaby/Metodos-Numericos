2. Explique, com suas palavras, os conceitos de vetorização e broadcasting que Numpy implementa. <br /> <br />
&emsp; Vetorização: Consiste em uma transformação linear que usa vetores e grandezas escalares <br />
&emsp; Broadcasting: Consiste em um método contínuo ou em cascata de operações entre os valores de arrays, ou seja, é uma soma vetorial! Então, se temos A = \[1, 2, 3] e B = \[1, 2, 3] a soma A + B = \[2, 4, 6]! Broadcasting só é aplicado a arrays de mesmo tamanho!

3. Liste as diferenças entre uma lista, um dicionário, um array numpy, dando exemplos de aplicação de em que cada tipo seria mais interessante. <br /> <br />
&emsp; A diferença está como a estrutura de dados se dá com os tipos dos valores que estão armazenado dentro delas e as operações que podem ser feitas! Na lista e no array numpy, os valores são armazenados da mesma forma, ou seja, são indexados, a diferença entre eles está nas operações que podem ser feitas, não podemos fazer vetorização, broadcasting e uppercasting forçado numa lista normal, operações que o array numpy tem suporte! O dicionário funciona com chaves e valores, um valor é "indexado" com uma chava, o dicionário também não suporta vetorização, broadcasting e uppercasting forçado!

6. Considere a lista \[-2, 4., ’7’, 9] e crie um array Numpy a partir dessa lista. Houve uma mudança no tipo de vários elementos. Explique com suas palavras porque isso aconteceu. Se o elemento ’7’ for escrito como inteiro, o que muda na saída e por que? <br /> <br />
&emsp; O que aconteceu foi é um método chamado uppercasting que faz um casting em um valor para um tipo que contenha uma prioridade maior! No caso da lista apresentada, temos valores que são string, int e float, desses 3 tipos, o string tem uma maior prioridade, logo os valores "não-string" viraram string! Caso eu o '7' vire um 7, o tipo com maior prioridade na lista será o float, logo, todos os elementos seram float!

10. Usando a função sin do numpy, e a função plot da matplotlib, módulo de geração de gráficos, plote um gráfico contendo os dois primeiros períodos de uma senóide.

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 4 * np.pi, 0.01)
y = np.sin(x)
plt.plot(x, y, 'r', label ='Vermelho')
plt.xlabel('Linha X')
plt.ylabel("Linha Y")
plt.title('Gráfico 1')
plt.legend()
plt.grid()
```
