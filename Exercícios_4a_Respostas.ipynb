1 Implemente o método da Bisseção descrito pelo algoritmo 1.1 no texto. Para testar o algoritmo,use os exemplos do material textual.

```python
# f = lambda x: 2 - np.exp(x)
f = lambda x: x ** 5 + 2 * x ** 3 - 5 * x - 2

def bissic(f, a, b, erro = 1e-5) :
    if (np.sign(f(a)) * np.sign(f(b)) != -1) :
        print("Não existe raiz nesse intervalo")
        return
    else :
        erroAt, erroAn = 1, a
        while (erroAt > erro) :
            x = (a + b) / 2
            print("a: %.8f b: %.8f x: %.8f" %(a, b, x))
            if (f(x) == 0) :
                print("Raiz encontrada! E ela é: ", x)
                return
            else :
                if (np.sign(f(a)) * np.sign(f(x)) == -1) :
                    b = x
                else :
                    a = x
                erroAt = np.abs(x - erroAn) / np.abs(x)
                print("erro: %.8f\n" %(erroAt))
                erroAn = x
        return
```

```python
bissic(f, -2, 2, 0.001)
```

```python
def raiz(a, b, c) :
    d = (b ** 2) - (4 * a * c)
    x1 = (-b + d ** (1 / 2)) / (2 * a)
    x2 = (-b - d ** (1 / 2)) / (2 * a)
    return x1, x2

raiz(-0.5, 2.5, 4.5)
```
